# Additions:
## 1. Adding borders
## 2. Adding subtitles

```python
from PIL import Image as pili, ImageOps as piliops
```

```python
from PIL import Image as pili, ImageDraw as pild, ImageFont as pilf 
from PIL import Image, ImageFont, ImageDraw
    
TINT_COLOR = (255, 255, 255)  # Black
TRANSPARENCY = .25  # Degree of transparency, 0-100%
OPACITY = int(155 * TRANSPARENCY)

a=0
while a<101:
    b=a+1
    if b==101:
        break
    print(b)
    piliops.expand(pili.open('data/avengers/talk/talk_{0}.jpg'.format(b)), border=(100,100),fill='black').save('data/avengers/modified/talk_{0}.jpg'.format(b))
    caption = dialog.values[a][0]
    print(caption)
    img = pili.open('data/avengers/modified/talk_{0}.jpg'.format(b)).convert('RGBA')
    overlay = pili.new('RGBA', img.size, TINT_COLOR+(0,))
    draw = pild.Draw(overlay)
    font = pilf.truetype("arial.ttf", 20)
    #x, y = (img.width - 510, img.height-100)
    text = caption
    w, h = font.getsize(text)
    num_lines = len(text.split('\n'))
    print(num_lines, w, h)
    x, y = 50, img.height - (num_lines-0.1*num_lines)*h
    #draw.rectangle((x, y, x + w, y + h), fill='black')
    #draw.rectangle((x, y, x + w, y + 4*h), fill=TINT_COLOR+(OPACITY,))
    draw.rectangle((x, y, x + img.width - 200, y + (num_lines-0.1*num_lines)*h), fill=TINT_COLOR+(OPACITY,))
    draw.text((x, y), text, fill=(209, 239, 8), font=font)

    # Alpha composite these two images together to obtain the desired result.
    img = pili.alpha_composite(img, overlay)
    img = img.convert("RGB") # Remove alpha for saving in jpg format.

    img.save('data/avengers/final/talk_{0}.jpg'.format(b))
    a += 1
```

# Horizontal Strip

```python
import numpy as np
import PIL  
```

```python

i=1
j=1
num=1
images=[]

while i<101:
    while j<6:
        if i == 101:
            break
        images.append('data/avengers/final/talk_{0}.jpg'.format(i))
        imgs = [ PIL.Image.open(i) for i in images ]
        
        # pick the image which is the smallest, and resize the others to match it (can be arbitrary image shape here)
        min_shape = sorted( [(np.sum(i.size), i.size ) for i in imgs])[0][1]
        imgs_comb = np.hstack( (np.asarray( i.resize(min_shape) ) for i in imgs ) )

        # save that beautiful picture
        imgs_comb = PIL.Image.fromarray( imgs_comb)
        
        if j==5:
            imgs_comb.save('data/avengers/horizontal/horizontal_{0}.jpg'.format(num))
            num+=1
#           imgs_comb.save('data/avengers/vertical/vertical_{0}.jpg'.format(i % 4)) 
            
        print(i,j)
        i+=1
        j+=1
        if j == 6:
            j=1
            images=[]
        
```

# Vertical strip

```python

i=1
j=1
numr=1
images=[]

while i<21:   ##i=1
    while j<5:  ##j=1,2
        if i==21:
            break
        
        images.append('data/avengers/horizontal/horizontal_{0}.jpg'.format(i))
        imgs = [ PIL.Image.open(i) for i in images ]
        
        # pick the image which is the smallest, and resize the others to match it (can be arbitrary image shape here)
        min_shape = sorted( [(np.sum(i.size), i.size ) for i in imgs])[0][1]
        imgs_comb = np.vstack( (np.asarray( i.resize(min_shape) ) for i in imgs ) )

        # save that beautiful picture
        imgs_comb = PIL.Image.fromarray( imgs_comb)
        
        if j==4:
            imgs_comb.save('data/avengers/vertical/vertical_{0}.jpg'.format(numr)) 
            numr += 1
        
        print(i,j)
        if j == 4:
            j=1
            i+=1
            images=[]
        else:    
            i+=1
            j+=1
            
    
```

# Convert to PDF

```python
from PIL import Image
from fpdf import FPDF

#pdf = FPDF()
f=1
imagelist=[]

while f<6:
    
    cover = Image.open('data/avengers/vertical/vertical_{0}.jpg'.format(1))
    width, height = cover.size
    pdf = FPDF(unit = "pt", format = [width, height])

    # imagelist is the list with all image filenames
    #mages.append('data/avengers/horizontal/horizontal_{0}.jpg'.format(i))
    imagelist.append('data/avengers/vertical/vertical_{0}.jpg'.format(f)) 
    if f== 5:
        for image in imagelist:
            pdf.add_page()
            pdf.image(image, 0, 0, width, height)
        pdf.output("data/avengers/avengers2.pdf", "F")
        print("data/avengers/avengers1111.pdf", "F")
        print("done!")
        break
    f+=1
```
